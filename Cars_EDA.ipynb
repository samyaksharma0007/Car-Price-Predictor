```python

print("DataFrame head with potential outliers in Price:")
print(cars.head())
print("\nDataFrame info to confirm 'Price' Dtype:")
cars.info()

# --- 1. Find Outliers using the IQR Method ---
print("\n--- Identifying Outliers in 'Price' column using IQR Method ---")

Q1_price = cars['price'].quantile(0.25)
Q3_price = cars['price'].quantile(0.75)
IQR_price = Q3_price - Q1_price

lower_bound_price = Q1_price - 1.5 * IQR_price
upper_bound_price = Q3_price + 1.5 * IQR_price

# Identify the outliers
outliers_price = cars[(cars['price'] < lower_bound_price) | (cars['price'] > upper_bound_price)]

print(f"\nQ1 (25th percentile): {Q1_price:,.2f}")
print(f"Q3 (75th percentile): {Q3_price:,.2f}")
print(f"IQR: {IQR_price:,.2f}")
print(f"Lower Bound for Outliers: {lower_bound_price:,.2f}")
print(f"Upper Bound for Outliers: {upper_bound_price:,.2f}")

print(f"\nTotal number of outliers found in 'Price' column: {len(outliers_price)}")

if not outliers_price.empty:
    print("\nDetails of Outliers in 'Price' column (first 10 if many):")
    print(outliers_price[['price']].head(10)) # Show just the price column for outliers
else:
    print("\nNo outliers found in the 'Price' column based on the IQR method.")

# --- 2. Visualize Outliers and Distribution ---

plt.figure(figsize=(15, 6))

# Subplot 1: Box Plot for Outlier Visualization
plt.subplot(1, 2, 1) # 1 row, 2 columns, 1st plot
sns.boxplot(y=cars['price'])
plt.title('Box Plot of Car Price (Outlier Visualization)')
plt.ylabel('Price')

# Subplot 2: Histogram for Distribution and Outlier Spread
plt.subplot(1, 2, 2) # 1 row, 2 columns, 2nd plot
sns.histplot(cars['price'], kde=True) # kde=True adds a Kernel Density Estimate line
plt.title('Histogram of Car Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')

plt.tight_layout()
plt.savefig('price_outliers_visualization.png')
plt.show()

# --- Optional: Violin Plot (combines box plot and KDE) ---
plt.figure(figsize=(8, 6))
sns.violinplot(y=cars['price'])
plt.title('Violin Plot of Car Price (Distribution & Outliers)')
plt.ylabel('Price')
plt.savefig('price_violin_plot.png')
plt.show()
```

## Extracting training data

```python
car = cars.copy()
```

```python
X=car[['name','company','year','kms_driven','fuel_type']]
y=car['Price']
```

```python
X
```

```python
y.shape
```

## Applying Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

## Creating an OneHotEncoder object to contain all the possible categories

```python
ohe=OneHotEncoder()
ohe.fit(X[['name','company','fuel_type']])
```

## Creating a column transformer to transform categorical columns

```python
column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),
                                    remainder='passthrough',force_int_remainder_cols=False)
```

## Linear Regression Model

```python
lr=LinearRegression()
```

## Making a pipeline

```python
pipe=make_pipeline(column_trans,lr)
```

## Fitting the model into the pipeline

```python
pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
```

## Checking R2 score

```python
r2_score(y_test,y_pred)
```

### Finding the model with a random state of train test split where the model gives max r2 score

```python
scores=[]
for i in range(2500):
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(X_train,y_train)
    y_pred=pipe.predict(X_test)
    scores.append(r2_score(y_test,y_pred))
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
pipe.predict(pd.DataFrame(columns=X_test.columns,data=np.array(['Toyota Innova Crysta 2.8','Toyota',2016,36000,'Diesel']).reshape(1,5)))
```

```python
cars[cars['company'] == 'Toyota']
```

## The best model is found at a certain random state

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=np.argmax(scores))
lr=LinearRegression()
pipe=make_pipeline(column_trans,lr)
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
r2_score(y_test,y_pred)
```
